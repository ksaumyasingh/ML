```python
#8. You have a dictionary d1 = {"k1":"v1","k2":"v2","k3":"v3","k4":"v4","k5":"v5"}. WAP to reverse this i.e keys should become values and values should become keys. The result should be d2 = {"v5":"k5","v4":"k4","v3":"k3","v2":"k2","v1":"k1"}
d1 = {"k1":"v1","k2":"v2","k3":"v3","k4":"v4","k5":"v5"}
print("Initial Data: ",str(d1))
inv_d1 ={v: k for k,v in d1.items()}
print("Final Data: ",dict(reversed(list(inv_d1.items()))))

# Output:
#     Initial Data:  {'k1': 'v1', 'k2': 'v2', 'k3': 'v3', 'k4': 'v4', 'k5': 'v5'}
#     Final Data:  {'v5': 'k5', 'v4': 'k4', 'v3': 'k3', 'v2': 'k2', 'v1': 'k1'}
```

```python
#9. WAP to input two dictionaries and print the overlapping keys in them.
d1 = {"k1":"v1","k2":"v2","k3":"v3","k4":"v4","k5":"v5"}
d2 = {"k6":"v6","k2":"v7","k7":"v7","k1":"v8","k9":"v9"}
l = []
for key in d1:
    if key in d2:
        l.append(key)
        
print("The overlapping Keys are:",l)
# Output:
#     The overlapping Keys are: ['k1', 'k2']
```

```python
#10. WAP to check if two same values in a dictionary has different keys.
d1 = {"k1":"v1","k2":"v2","k3":"v2","k4":"v4","k5":"v1"}
count = 0
val = list (d1.values())
for v1 in val :
    answer = val.count(v1)
    if answer > 1 :
        count += answer
    for v2 in val :
        if v1 == v2 :
            val.remove( v2 )
if count == 0 :
    print( "No values are same" )
else :
    print( count,"values are same")
```
