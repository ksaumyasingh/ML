```python
import pandas as pd
df=pd.read_csv("../input/insurance-dataset/insurance.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
categorical_names=['sex','smoker','region','children']
df_encoder=pd.get_dummies(data=df, prefix='OHE', prefix_sep='_', columns=categorical_names, drop_first=True, dtype='int8')
print(df)
print(df_encoder)
```

```python
X=df_encoder.drop(['charges'],axis=1)
y=df_encoder['charges']
print(X)
print(y)
```

```python
print(X.shape,y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_new=scaler.fit_transform(X)
X_new
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_new,y,test_size=0.2,random_state=49)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
df=pd.DataFrame({'Real values':y_test,'predicted values':y_pred})
df
```

```python
from sklearn.tree import DecisionTreeRegressor
model_DT=DecisionTreeRegressor().fit(X_train,y_train)
```

```python
y_pred_DT=model_DT.predict(X_test)
```

```python
df=pd.DataFrame({'Real values':y_test,'Predicted values':y_pred_DT})
df
```

```python
from sklearn.datasets import load_iris
X,y=load_iris(return_X_y=True)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=49)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier().fit(X_train,y_train)

```

```python
y_pred_DTC=model.predict(X_test)
```

```python
df=pd.DataFrame({'Real values':y_test,'Predicted values':y_pred_DTC})
df
```
