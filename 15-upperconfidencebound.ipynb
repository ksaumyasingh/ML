```python
import pandas as pd
data=pd.read_csv("../input/adoptimisation/Ads_Optimisation (1).csv")
data
```

```python
#random Selection
import random

total_reward=0
ads_selected=[]
N=10000
d=10

for i in range(0,N):
    ad=random.randrange(d)
    ads_selected.append(ad)
    reward=data.values[i,ad]
    total_reward+=reward
    
    
```

```python
import math
N=10000
d=10

ads_selected=[]
total_reward=0
no_of_selection=[0]*d
reward=[0]*d

for n in range(0,N):
    ad=0
    max_upper_bound=0
    for i in range(0,d):
        if no_of_selection[i]>0:
            avg_reward=reward[i]/no_of_selection[i]
            ucb=math.sqrt(2*math.log(n+1)/no_of_selection[i])
            upper_bound=avg_reward+ucb
        else:
            upper_bound=1e500
        if upper_bound>max_upper_bound:
            max_upper_bound=upper_bound
            ad=i
    ads_selected.append(ad)
    no_of_selection[ad]+=1
    rewards=data.values[n,ad]
    reward[ad]+=rewards
    total_reward+=rewards
        
        
        
            
```
