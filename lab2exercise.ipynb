```python
n = int(input("Enter the number to check : "))
sum = 0
temp = n
while temp > 0:
    num = temp % 10
    sum += num ** 3
    temp //= 10

if n == sum:
    print(n,"is an Armstrong number")
else:
    print(n,"is not an Armstrong number")
```

```python
import math
n = int(input("Enter the number to check for perfect square : "))
if(n >= 0):
        sr = int(math.sqrt(n))
        print(sr)
        if (sr*sr == (n)):
            print(n,"is a Perfect Square")
        else:
            print(n,"is not a Perfect Square")
```

```python
n = int(input("Enter the number of which you want to find the factorial : "))
fac = 1
if n == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1, n+1):
        fac = fac*i
print("The factorial of",n,"is",fac)
```

```python
import math
a = int(input("Enter coefficient of x^2 : "))
b = int(input("Enter the coefficient of x : "))
c = int(input("Enter the constant : "))

d = b * b - 4 * a * c
sqrtd = math.sqrt(d)

if d > 0:
    print("Roots are real and different ")
    print((-b + sqrtd)/(2 * a))
    print((-b - sqrtd)/(2 * a))
elif d == 0:
    print("Roots are real and same")
    print(-b / (2*a))
else:
    print("Roots are complex")
    print(- b / (2*a), " + i", sqrtd)
    print(- b / (2*a), " - i", sqrtd)
```
