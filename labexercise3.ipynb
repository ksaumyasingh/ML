```python
str = input("Enter string:")
length = len(str)
maxlength = 0
maxsub =''              
sub =''                 
lensub = 0
for i in range(length):
    if (str[i] in 'aeiou' or str[i] in 'AEIOU'):
         maxsub = sub
         maxlength = lensub
         sub = ''
         lensub = 0
    else:
        sub += str[i]
        lensub = len(sub)
        i += 1
print(maxsub)
```

```python
str = input("Enter the string:")
length = len(str)
i = 0
ans = ""
while(i < length):
    if(i%2 != 0):
        ans += str[i].upper()
    else:
        ans += str[i]
    i += 1
print(ans)
    
```

```python
str = "This is Banasthali"
ss = str.lower()
v = 0
c = 0
for i in ss:
    if(i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u'):
        v += 1
    elif(i == ' '):
        continue
    else:
        c += 1
print(v)
print(c)
        
        
```

```python
##WAP to read a sentnece and count the articles it has
str = "I saw an accident of a car."
ss = (str.lower()).split()
cnt = 0
for i in ss:
    if(i == 'an' or i == 'a' or i == 'the'):
        cnt += 1
    else:
        continue
print(cnt)
    
```
