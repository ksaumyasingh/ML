```python
import pandas as pd
df=pd.read_csv('../input/wholesale/Wholesale customers data.csv')
df
```

```python
from sklearn.preprocessing import normalize
data=normalize(df)
data=pd.DataFrame(data,columns=df.columns)
data
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

fig=plt.figure(figsize=(10,10))
plt.title('Dendrogram')
dendo=sch.dendrogram(sch.linkage(data,method='ward'))
plt.axhline(y=6,c='red',linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
aglo=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
aglo.fit_predict(data)
```

```python
fig=plt.figure(figsize=(10,10))
plt.scatter(data['Milk'],data['Grocery'],c=aglo.labels_)
```
