```python
import pandas as pd
from sklearn.datasets import load_iris
data=load_iris()
data
```

```python
X,y=load_iris(return_X_y=True)
X
```

```python
y
```

```python
from xgboost import XGBClassifier
```

```python
model=XGBClassifier()
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=.2,random_state=47)
```

```python
model.fit(Xtrain,ytrain)
```

```python
ypred=model.predict(Xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred,target_names=data.target_names))
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(ypred,ytest)
print("Accuracy=%.2f%%" %(acc*100))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(ypred,ytest))
```

```python
labels=data.target_names
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm=confusion_matrix(ypred,ytest)
fig=plt.figure(figsize=(6,4))
sns.heatmap(cm,xticklabels=labels,yticklabels=labels,annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Values',fontweight='bold',fontsize=15)
plt.ylabel('True Values',fontweight='bold',fontsize=15)
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(model,Xtest, ytest, cv=5)
score
```
