```python
# Prepocessing techniques
# Involves formattting and making data clean for better performance of model
# Missing values, Encoding Categorical data: onehot encoder using dummy variables(0 or 1),
# Feature scaling: rescale data to specific range, includes standard scaler, minmaxscaler etc


from sklearn.impute import SimpleImputer
import numpy as np
X=[[1,2,3],[4,np.NaN,5],[6,7,8]]
imp=SimpleImputer(missing_values=np.NaN,strategy='mean')
print(imp.fit_transform(X))
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
print(iris)

```

```python
X,y=load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

```python
# data get rescaled such that mean=0 and standard deviation=1
from sklearn.preprocessing import StandardScaler  
scaler=StandardScaler()
print(scaler.fit(X))
print(scaler.transform(X))
```

```python
# data get rescaled between 0-1 range
from sklearn.preprocessing import MinMaxScaler 
scaler=MinMaxScaler()
print(scaler.fit_transform(X))
```

```python
from sklearn.preprocessing import MaxAbsScaler
scaler=MaxAbsScaler()
print(scaler.fit_transform(X))
```

```python
##transform feature to follow normal distribution & reduces outliers
from sklearn.preprocessing import QuantileTransformer 
scaler=QuantileTransformer()
print(scaler.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=47)
print(X_train,y_train)
print(X_test,y_test)
```

```python
from sklearn.preprocessing import OneHotEncoder
from numpy import asarray
x=asarray([['male','chrome','from UK',4],['female','safari','from India',5],['female','Opera','from US',6]])
encoder=OneHotEncoder()
output=encoder.fit_transform(x)
print(output)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from numpy import asarray
x=asarray([['male','chrome','from UK',4],['female','safari','from India',5],['female','Opera','from US',6]])
encoder=OrdinalEncoder()
output=encoder.fit_transform(x)
print(output)
```
