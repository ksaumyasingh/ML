```python
# There are 3 supervised classification techniques
# 1. decision tree
# 2. support vector machine
# 3. naive bayes
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
data
```

```python
import pandas as pd
X=pd.DataFrame(load_breast_cancer().data,columns=data.feature_names)
print(X)
print(type(X))
```

```python
y= load_breast_cancer().target
#y=pd.DataFrame(y)
print(y)
print(type(y))
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
X
```

```python
X.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
sfs=SFS(estimator=model,k_features=(10,20),forward=True,scoring='accuracy',cv=5)
sfs=sfs.fit(X,y)
```

```python
x_sfs=sfs.transform(X)
x_sfs
```

```python
x_sfs.shape
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(x_sfs,y,test_size=.2,random_state=47)
```

```python
model.fit(Xtrain,ytrain)
```

```python
ypred=model.predict(Xtest)
```

```python
print(model.score(Xtrain,ytrain))
print(model.score(Xtest,ytest)) #overfitting issue
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred,target_names=data.target_names))
```

```python
#naive bayes classification technique
from sklearn.naive_bayes import GaussianNB
model_nb=GaussianNB()
```

```python
model_nb.fit(Xtrain,ytrain)
```

```python
ypred1=model_nb.predict(Xtest)
```

```python
print(model_nb.score(Xtrain,ytrain))
print(model_nb.score(Xtest,ytest))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred1,target_names=data.target_names))
```

```python
# Support vector machine classification algorithm
from sklearn import svm
model_svc=svm.SVC()
```

```python
model_svc.fit(Xtrain,ytrain)
```

```python
ypred2=model_svc.predict(Xtest)
```

```python
print(model_svc.score(Xtrain,ytrain))
print(model_svc.score(Xtest,ytest))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred2,target_names=data.target_names))
```
