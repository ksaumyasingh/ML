```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
X=df.drop('target',axis=1)
X
```

```python
y=df.target
type(y)
```

```python
y=pd.DataFrame(y)
y
```

```python
fs=SelectKBest(chi2,k=2)
fs=fs.fit(X,y)
```

```python
x_best=fs.transform(X)
x_best
```

```python
column=fs.get_support(indices=True)
column
```

```python
new_features=X.iloc[:,column]
new_features
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(new_features,y,test_size=.3,random_state=47)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier();
model.fit(Xtrain,ytrain)
```

```python
y_pred=model.predict(Xtest)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,y_pred,target_names=data.target_names))
```
