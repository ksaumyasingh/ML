```python
X,y=load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,shuffle=True)
for train,test in kf.split(X,y):
    Xtrain,Xtest=X[train],X[test]
    ytrain,ytest=y[train],y[test]
        

```

```python
Xtest
```

```python
from sklearn.linear_model import LinearRegression
model_lr=LinearRegression().fit(Xtrain,ytrain)
```

```python
y_pred_lr=model_lr.predict(Xtest)
y_pred_lr
```

```python
from sklearn import metrics
import numpy as np
print("Mean Absolute Error= ",metrics.mean_absolute_error(ytest,y_pred_lr))
print("Mean Squared Error= ",metrics.mean_squared_error(ytest,y_pred_lr))
print("Root Mean Square Error= ",np.sqrt(metrics.mean_squared_error(ytest,y_pred_lr)))
```

```python
model_lr.score(Xtrain,ytrain)
```

```python
model_lr.score(Xtest,ytest)
```

```python
from sklearn.tree import DecisionTreeRegressor
model_dtr=DecisionTreeRegressor().fit(Xtrain,ytrain)
```

```python
y_pred_dtr=model_dtr.predict(Xtest)
y_pred_dtr
```

```python
from sklearn import metrics
import numpy as np
print("Mean Absolute Error= ",metrics.mean_absolute_error(ytest,y_pred_dtr))
print("Mean Squared Error= ",metrics.mean_squared_error(ytest,y_pred_dtr))
print("Root Mean Square Error= ",np.sqrt(metrics.mean_squared_error(ytest,y_pred_dtr)))
```

```python
model_dtr.score(Xtest,ytest)
```

```python
from sklearn.ensemble import StackingRegressor
level0=list()
level0.append(('lr',LinearRegression()))
level0.append(('dt',DecisionTreeRegressor()))
level1=LinearRegression()
model_en=StackingRegressor(estimators=level0,final_estimator=level1,cv=5)
```

```python
model_en.fit(Xtrain,ytrain)
```

```python
y_pred_en=model_en.predict(Xtest)
```

```python
from sklearn import metrics
import numpy as np
print("Mean Absolute Error= ",metrics.mean_absolute_error(ytest,y_pred_en))
print("Mean Squared Error= ",metrics.mean_squared_error(ytest,y_pred_en))
print("Root Mean Square Error= ",np.sqrt(metrics.mean_squared_error(ytest,y_pred_en)))
```

```python
model_en.score(Xtest,ytest)
```

```python
model_en.score(Xtrain,ytrain)
```
