# Creating a Matrix

```python
R = int(input("Enter the number of rows:")) 
C = int(input("Enter the number of columns:")) 
  
# Initialize matrix 
matrix = [] 
print("Enter the entries rowwise:") 
  
# For user input 
for i in range(R):          # A for loop for row entries 
    a =[] 
    for j in range(C):      # A for loop for column entries 
         a.append(int(input())) 
    matrix.append(a) 
  
# For printing the matrix 
for i in range(R): 
    for j in range(C): 
        print(matrix[i][j], end = " ") 
    print() 
```

```python
# 8. WAP to input a list of numbers and shift all zeros to the right and all non zeros to left of the list

lst = []            
num = int(input("Enter the size of list : "))
for i in range(num):
    ele = int(input()) 
    lst.append(ele)
count = 0
for i in range(num):
    if lst[i] != 0:
        lst[count] = lst[i]
        count+=1
while count < num:
    lst[count] = 0
    count += 1
    
print("List after pushing all zeros to end of list:")
print(lst)
```

```python
# 9. WAP to add two matrices
X = [[1,2,3],
    [4 ,5,6],
    [7 ,8,9]]
  
Y = [[9,18,7],
    [6,5,4],
    [3,2,1]]
 
result = [[X[i][j] + Y[i][j]  for j in range(len(X[0]))] for i in range(len(X))]
  
for r in result:
    print(r)
```

```python
# 12. WAP to add elements of diagonal of a matrix

mat = [[ 1, 2, 3, 9 ],
     [ 5, 6, 7, 8 ],
     [ 1, 2, 3, 4 ],
     [ 5, 6, 7, 8 ]]

principal = 0
secondary = 0

for i in range(len(mat)):
    principal += mat[i][i]
    secondary += mat[i][len(mat) - i - 1]
         
print("Principal Diagonal:", principal)
print("Secondary Diagonal:", secondary)
```

```python
# 13. WAP to find an element in a matrix

test_list = [[4, 5, 6],
             [10, 2, 13],
             [1, 11, 18]]
  
print("The original list : " + str(test_list))
  
num = int(input("Enter the element you want to search : "))    
res = any(num in sub for sub in test_list)
if res == True:
    print(num," is present in matrix")
else:
    print(num," is not present in matrix")
```
